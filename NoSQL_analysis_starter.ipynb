```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
uk_food = mongo['uk_food']
```

```python
# review the collections in our database
print(uk_food.list_collection_names())
```

```python
# assign the collection to a variable
establishments = uk_food['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": { "$eq": 20 }}

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query))
```

```python
# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_establishments_df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
len(hygiene_establishments_df)

# Display the first 10 rows of the DataFrame
hygiene_establishments_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
rating_list = ['4', '5', 'Pass']
query = {"LocalAuthorityName": {'$regex':'London' }, "RatingValue": {'$in': rating_list}}

# Use count_documents to display the number of documents in the result
establishments.count_documents(query)
```

```python
# Display the first document in the results using pprint
establishments.find_one(query)
```

```python
# Convert the result to a Pandas DataFrame
ratingvalue_establishments_df = pd.DataFrame(list(establishments.find(query)))

# Display the number of rows in the DataFrame
len(ratingvalue_establishments_df)

# Display the first 10 rows of the DataFrame
ratingvalue_establishments_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
results = establishments.find({"BusinessName" : "Penang Flavours"}, {"geocode": 1})
{'car_year': {'$gte': 2010}}
for result in results:
    latitude_upper = result["geocode"]["latitude"] + degree_search
    latitude_lower = result["geocode"]["latitude"] - degree_search
    longitude_upper = result["geocode"]["longitude"] + degree_search
    longitude_lower = result["geocode"]["longitude"] - degree_search


query = {"geocode.latitude" : {"$gte": latitude_lower, "$lte" :  latitude_upper},
         "geocode.longitude" : {"$gte": longitude_lower, "$lte" :  longitude_upper},
         "RatingValue" : { "$in": ['5']} }
limit = 5
sort = [("scores.Hygiene", 1)]

# Print the results
results = establishments.find(query).sort(sort).limit(limit)
for result in results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
nearest_establishments_df = pd.DataFrame(list(establishments.find(query)))
nearest_establishments_df
```
